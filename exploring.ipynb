```python
import PIL.Image as Image
img_path = "./data/training/0000/000000.png"
img = np.array(Image.open(img_path))
obj_ids = np.unique(img)
# to correctly interpret the id of a single object
obj_id = obj_ids[0]
print(obj_id)
class_id = obj_id // 1000
print(class_id)
obj_instance_id = obj_id % 1000
print(obj_instance_id)
```

```python
ann_filename = "./data/instances/0000/000000.png"
ann_data = tf.io.read_file(ann_filename)
ann = tf.image.decode_image(ann_data, dtype=tf.uint16, channels=1)
```

```python
# Plot the image and annotations side by side
plt.figure(figsize=(10, 5))

# Plot the image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Image')
plt.axis('off')

# Plot the annotations
plt.subplot(1, 2, 2)
plt.imshow(ann[:, :, 0], cmap='gray')
plt.title('Annotations')
plt.axis('off')

plt.show()
```
